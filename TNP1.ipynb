```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Sum:", a + b)
print("Difference:", a - b)
print("Product:", a * b)
print("Division:", a / b)

print("Even/Odd:", "Even" if a % 2 == 0 else "Odd")

print("Float value:", float(a))
```

```python
s = input("Enter sentence: ")

vowels = "aeiouAEIOU"
v = sum(1 for ch in s if ch in vowels)
c = sum(1 for ch in s if ch.isalpha() and ch not in vowels)

print("Vowels:", v)
print("Consonants:", c)
print("Reverse:", s[::-1])
print("Underscore:", s.replace(" ", "_"))
print("Capitalized:", s.title())
```

```python
t = (1, "a", 5, 10.5, 3)

nums = [x for x in t if type(x) in [int, float]]
print("Numbers:", nums)

# Tuple can't be modified
try:
    t[0] = 100
except:
    print("Cannot modify tuple")

t2 = (9, 8)
print("Concatenated:", t + t2)
```

```python
students = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Name: ")
    marks = int(input("Marks: "))
    students[name] = marks

avg = sum(students.values()) / len(students)
topper = max(students, key=students.get)

print("Average:", avg)
print("Topper:", topper)
```

```python
contacts = {}

while True:
    ch = input("1.Add 2.Search 3.Delete 4.Show 5.Exit: ")

    if ch == "1":
        name = input("Name: ")
        num = input("Number: ")
        contacts[name] = num
    elif ch == "2":
        name = input("Search: ")
        print(contacts.get(name, "Not found"))
    elif ch == "3":
        name = input("Delete: ")
        contacts.pop(name, None)
    elif ch == "4":
        print(contacts)
    else:
        break
```

```python
emp = {}

while True:
    ch = input("1.Add 2.Remove 3.Show 4.Exit: ")

    if ch == "1":
        name = input("Name: ")
        emp[name] = "Present"
    elif ch == "2":
        name = input("Remove: ")
        emp.pop(name, None)
    elif ch == "3":
        print(emp)
    else:
        break
```

```python
students = []

n = int(input("Enter count: "))

for i in range(n):
    name = input("Name: ")
    roll = int(input("Roll: "))
    marks = int(input("Marks: "))
    students.append({"name": name, "roll": roll, "marks": marks})

for s in students:
    if s["marks"] >= 40:
        print(s["name"], "Pass")
    else:
        print(s["name"], "Fail")
```
