```python
class Library:
    def __init__(self):
        self.books = {}

    def add(self, name, qty):
        self.books[name] = self.books.get(name, 0) + qty

    def view(self):
        if not self.books:
            print("No books available")
        else:
            for b, q in self.books.items():
                print(b, "-", q)

    def issue(self, name):
        if self.books.get(name, 0) > 0:
            self.books[name] -= 1
            print("Book issued")
        else:
            print("Not available")

    def return_book(self, name):
        self.books[name] = self.books.get(name, 0) + 1
        print("Book returned")


lib = Library()

while True:
    print("\n1.Add  2.View  3.Issue  4.Return  5.Exit")
    ch = input("Choice: ")

    if ch == "1":
        lib.add(input("Book name: "), int(input("Qty: ")))
    elif ch == "2":
        lib.view()
    elif ch == "3":
        lib.issue(input("Book name: "))
    elif ch == "4":
        lib.return_book(input("Book name: "))
    elif ch == "5":
        break
    else:
        print("Invalid choice")
```

```python
class Cart:
    def __init__(self):
        self.items = {}

    def add(self, name, price, qty):
        self.items[name] = [price, qty]

    def remove(self, name):
        if name in self.items:
            del self.items[name]
            print("Item removed")
        else:
            print("Item not found")

    def view(self):
        total = 0
        if not self.items:
            print("Cart is empty")
        else:
            for name, (price, qty) in self.items.items():
                cost = price * qty
                total += cost
                print(name, "-", qty, "x", price, "=", cost)
            print("Total:", total)


cart = Cart()

while True:
    print("\n1.Add  2.Remove  3.View  4.Exit")
    ch = input("Choice: ")

    if ch == "1":
        name = input("Product name: ")
        price = float(input("Price: "))
        qty = int(input("Quantity: "))
        cart.add(name, price, qty)

    elif ch == "2":
        cart.remove(input("Product name: "))

    elif ch == "3":
        cart.view()

    elif ch == "4":
        break

    else:
        print("Invalid choice")
```

```python
class ATM:
    def __init__(self):
        self.accounts = {
            "1001": {"pin": "1234", "balance": 5000},
            "1002": {"pin": "5678", "balance": 8000}
        }

    def login(self, acc, pin):
        if acc in self.accounts and self.accounts[acc]["pin"] == pin:
            return True
        return False

    def check_balance(self, acc):
        print("Balance:", self.accounts[acc]["balance"])

    def deposit(self, acc, amt):
        self.accounts[acc]["balance"] += amt
        print("Deposited Successfully")

    def withdraw(self, acc, amt):
        if self.accounts[acc]["balance"] >= amt:
            self.accounts[acc]["balance"] -= amt
            print("Withdraw Successful")
        else:
            print("Insufficient Balance")


atm = ATM()

acc = input("Enter Account Number: ")
pin = input("Enter PIN: ")

if atm.login(acc, pin):
    while True:
        print("\n1.Balance  2.Deposit  3.Withdraw  4.Exit")
        ch = input("Choice: ")

        if ch == "1":
            atm.check_balance(acc)
        elif ch == "2":
            atm.deposit(acc, float(input("Amount: ")))
        elif ch == "3":
            atm.withdraw(acc, float(input("Amount: ")))
        elif ch == "4":
            break
        else:
            print("Invalid choice")
else:
    print("Invalid Account or PIN")
```

```python
class GradeBook:
    def __init__(self):
        self.students = {}

    def add_student(self, name, marks):
        self.students[name] = marks

    def view_students(self):
        if not self.students:
            print("No records found")
        else:
            for name, marks in self.students.items():
                avg = sum(marks) / len(marks)
                print(name, "Marks:", marks, "Average:", round(avg, 2))

    def calculate_grade(self, avg):
        if avg >= 90:
            return "A"
        elif avg >= 75:
            return "B"
        elif avg >= 60:
            return "C"
        elif avg >= 40:
            return "D"
        else:
            return "F"


gb = GradeBook()

while True:
    print("\n1.Add Student  2.View Records  3.Exit")
    ch = input("Choice: ")

    if ch == "1":
        name = input("Student Name: ")
        marks = list(map(int, input("Enter marks separated by space: ").split()))
        gb.add_student(name, marks)

    elif ch == "2":
        if not gb.students:
            print("No records found")
        else:
            for name, marks in gb.students.items():
                avg = sum(marks) / len(marks)
                grade = gb.calculate_grade(avg)
                print(name, "Average:", round(avg, 2), "Grade:", grade)

    elif ch == "3":
        break
    else:
        print("Invalid choice")
```

```python
import hashlib

class PasswordManager:
    def __init__(self):
        self.users = {}

    def hash_password(self, password):
        return hashlib.sha256(password.encode()).hexdigest()

    def register(self, username, password):
        if username in self.users:
            print("User already exists")
        else:
            self.users[username] = self.hash_password(password)
            print("User registered successfully")

    def login(self, username, password):
        if username in self.users:
            if self.users[username] == self.hash_password(password):
                print("Login successful")
            else:
                print("Incorrect password")
        else:
            print("User not found")


pm = PasswordManager()

while True:
    print("\n1.Register  2.Login  3.Exit")
    ch = input("Choice: ")

    if ch == "1":
        u = input("Username: ")
        p = input("Password: ")
        pm.register(u, p)

    elif ch == "2":
        u = input("Username: ")
        p = input("Password: ")
        pm.login(u, p)

    elif ch == "3":
        break

    else:
        print("Invalid choice")
```
