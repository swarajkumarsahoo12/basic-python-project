```python
class ParkingLot:
    def __init__(self, capacity):
        self.capacity = capacity
        self.vehicles = []

    def entry(self, vehicle):
        if len(self.vehicles) < self.capacity:
            self.vehicles.append(vehicle)
            print("Vehicle Entered:", vehicle)
        else:
            print("Parking Full")

    def exit(self, vehicle):
        if vehicle in self.vehicles:
            self.vehicles.remove(vehicle)
            print("Vehicle Exited:", vehicle)
        else:
            print("Vehicle not found")

    def available_spots(self):
        print("Available Spots:", self.capacity - len(self.vehicles))

p = ParkingLot(5)
p.entry("KA01AB1234")
p.entry("KA02CD5678")
p.available_spots()
p.exit("KA01AB1234")
p.available_spots()
```

```python
tasks = []

def add_task(name, category, priority):
    tasks.append({"task": name, "category": category, "priority": priority})

def show_tasks():
    for t in tasks:
        print(t)

add_task("Finish Assignment", "Study", "High")
add_task("Buy Groceries", "Personal", "Medium")

show_tasks()
```

```python
products = {
    "pen": 10,
    "notebook": 50,
    "eraser": 5
}

bill = 0

while True:
    item = input("Enter product (or 'done'): ")
    if item == "done":
        break

    qty = int(input("Quantity: "))
    cost = products[item] * qty
    bill += cost

print("Total Bill:", bill)
```

```python
seats = 10

while True:
    print("Available seats:", seats)
    n = int(input("Tickets to book: "))

    if n <= seats:
        seats -= n
        print("Booking Confirmed")
    else:
        print("Not enough seats")

    if seats == 0:
        print("House Full")
        break
```

```python
students = {}

def add_student(name, marks):
    students[name] = marks

def calculate_gpa(marks):
    return sum(marks)/len(marks)

add_student("Rahul",[80,75,90])
add_student("Anita",[70,85,88])

for s in students:
    gpa = calculate_gpa(students[s])
    print(s,"GPA:",round(gpa/10,2))
```

```python
passwords = {}

while True:
    print("\n1.Add Password 2.View Password 3.Exit")
    ch = int(input("Choice: "))

    if ch == 1:
        site = input("Website: ")
        pwd = input("Password: ")
        passwords[site] = pwd

    elif ch == 2:
        site = input("Website: ")
        print("Password:", passwords.get(site, "Not found"))

    elif ch == 3:
        break
```

```python
freelancers = []
clients = []

def register_freelancer(name, skill):
    freelancers.append((name,skill))

def register_client(name, project):
    clients.append((name,project))

register_freelancer("John","Web Development")
register_client("CompanyA","Website")

print("Freelancers:",freelancers)
print("Clients:",clients)
```

```python
inventory = {}

def add_item(item, qty):
    inventory[item] = inventory.get(item,0) + qty

def remove_item(item, qty):
    if item in inventory and inventory[item] >= qty:
        inventory[item] -= qty

def show_inventory():
    print(inventory)

add_item("Laptop",10)
add_item("Mouse",50)
remove_item("Mouse",5)

show_inventory()
```
